```python
import pandas as pd
import numpy as np
```

```python
sales = pd.read_json("../reviews_Electronics_5.json", lines=True)
```

```python
sales = sales[['reviewerID', 'asin', 'overall']]
sales
```

```python
asin_sales = sales.groupby(by=['asin'], dropna=False).count()
asin_sales = asin_sales[asin_sales.apply(lambda x: x > 500)].dropna().drop('overall', axis=1).rename(columns = {'reviewerID': 'review_count'})
asin_sales
```

```python
sales = sales.assign(over500=sales.asin.isin(asin_sales.index).astype(str))
sales
```

```python
sales = sales[sales['over500'] == "True"].drop('over500', axis = 1)
sales
```

```python
sales_pivot = sales.pivot_table(index='reviewerID', columns='asin', values='overall', fill_value=0)
sales_pivot
```

```python
X = sales_pivot.T
X_index = X.index
```

```python
X
```

```python
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=100)
decomposed_matrix = SVD.fit_transform(X)
decomposed_matrix
```

```python
correlation_matrix = np.corrcoef(decomposed_matrix)
corr_mat = pd.DataFrame(correlation_matrix, index=X.index, columns = X.index)
corr_mat.head()
```

```python
corr_mat.to_csv("RevisedCorr.csv")
```

```python
i = str(X.index[75])
product_names = list(X.index)
product_ID = product_names.index(i)

correlation_product_ID = correlation_matrix[product_ID]

Recommend = list(X.index[correlation_product_ID > 0.1])
Recommend.remove(i)
```

```python
Recommend
```

```python
indexes = X.index.tolist()
with open('product_names.txt', 'w') as f:
    f.write('\n'.join(indexes))
```
