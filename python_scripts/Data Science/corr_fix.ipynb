```python
import numpy as np
import pandas as pd
import json
```

```python
with open('./product_details.json', 'r') as product_details:
    prod_details = json.load(product_details)

prod_details[:5]
```

```python
corr_mat = pd.read_csv('RevisedCorr.csv', index_col=0)
corr_mat.head()
```

```python
available_products = [prod['asin'] for prod in prod_details]

unavailable_products = []

for prod in corr_mat.index:
    if prod not in available_products:
        unavailable_products.append(prod)
        
unavailable_products
```

```python
for oos_asin in unavailable_products:
    corr_mat.drop(oos_asin, inplace=True)
    corr_mat.drop(oos_asin, axis=1, inplace=True)

corr_mat
```

```python
corr_mat.to_csv('FinalCorr.csv')
```
